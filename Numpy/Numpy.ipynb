### **Arrays**
**Tipos de Arrays (ndarrays)** -> Arrays com N dimensões.

**1D Array** -> Uma dimensão, chamado de vetor ou vector.

**2D Array** -> Duas dimensões, chamado de matriz ou matrix.

**3D ou + Array** -> Três dimensões ou mais, chamado de tensor.


### **Funções Importantes**
```
import numpy as np
```

**np.array([])** -> Transforma listas em um Array Numpy.

**np.arange(i, f, p)** -> Cria um Array a partir de um Range de Números.

**np.linspace(start, stop, qtd=n)** -> Cria um Array seguindo uma sequência linear (n = qtd de números).

**np.loadtxt(path ou url, delimiter= ',', usecols= range)** -> Transformar arquivo CSV em Array.

**np.concatenate((a, b)**) -> Concatenar dois Arrays.

**np.array_equal(x, y)** -> Verifica se os Arrays são iguais.

**np.allclose(x, y, n)** -> Verifica se tem diferença menor que (n) entre os Arrays.

**np.isnan(array)** -> Verifica os valores NaNs do Array, dá pra fazer a sum(np.isnan(array)) para ver a quantidade.

**np.mean(array)** -> Calcula a média dos valores do Array.

**np.random.radint(low=1, high=100, size= 40)** -> Gerar 40 números inteiros aleatórios entre 1 e 100.

**np.random.uniform(low=0.1, high=1, size= 50)** -> Gerar 50 números floats aleatórios entre 0.1 e 1.

**np.random.seed(n)** -> Acessa uma seed de números aleatórios (n) para reprodutibilidade. 

**np.column_stack([array1, array2])** -> Une arrays 1D em um array 2D.

**np.savetxt('nome_do_arquivo', array, delimiter= ',')** -> Salva o Array em um arquivo .csv

### **Métodos Importantes**

**array.shape** -> Mostra a quantidade de Linhas e Colunas, respectivamente.

**array.ndim** -> Mostra o número de dimensões da matriz (Linha e Coluna = 2 dimensões).

**array.T** -> Faz a transposição da matriz.

**array.size** -> Mostra a quantidade de dados na matriz (parece com a função len()).

**array[lógica]** -> Separar valores do Array que seguem uma lógica. Ex: array[array > 10] -> Separa somente os valores maiores que 10.

**array.sum, max, min, mean()** -> Operações com o array.

````python
import numpy as np
import time

# cria uma lista com 1000000 elementos
lista = list(range(1000000))

# transforma a lista em um array Numpy
array = np.array(lista)

# começa a cronometrar o tempo para a operação com a lista
start_time = time.time()

# realiza a operação de elevar ao quadrado cada elemento da lista
lista_quadrado = [i**2 for i in lista]

# para o cronômetro
list_time = time.time() - start_time

# começa a cronometrar o tempo para a operação com o array
start_time = time.time()

# realiza a operação de elevar ao quadrado cada elemento do array
array_quadrado = array**2

# para o cronômetro
array_time = time.time() - start_time

print("Tempo da operação com a lista: ", list_time)
print("Tempo da operação com o array: ", array_time)
````
