```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv')
```

```python
df.set_index('Pa칤s', inplace=True)
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
```

```python
sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h');
```

```python
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set(title='Pa칤ses com maior imigra칞칚o para o Canad치\n1980 a 2013',
       xlabel='N칰mero de imigrantes',
       ylabel='')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Pa칤ses com maior imigra칞칚o para o Canad치\n1980  2013', fontsize=16, loc='left')
ax.set_xlabel('N칰mero de Imigrantes', fontsize=14)
ax.set_ylabel('')
plt.show()
```

```python
# Fun칞칚o para gerar paletas diferentes
def gerar_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
    ax.set_title('Pa칤ses com maior imigra칞칚o para o Canad치\n1980  2013', fontsize=16)
    ax.set_xlabel('N칰mero de Imigrantes', fontsize=14)
    ax.set_ylabel('')
    
    plt.show()
```

```python
gerar_paleta('Blues_r')
```

```python
gerar_paleta('tab10')
```

```python
gerar_paleta('rocket')
```
