**pd.DataFrame()** -> Transformar alguma variável em um DataFrame

**pd.Series()** -> Transformar alguma variável em uma Series

>### **Métodos Importantes**

**dados.info()** -> Mostra diversas informações estruturais do DataFrame 

**dados.columns()** -> Mostra as colunas do DataFrame 

**dados.head(n) ou dados.tail(n)** -> Mostra as primeiras ou ultimas linhas do DataFrame

**dados['Nome da Coluna'] ou dados[['Coluna1', 'Coluna2']]** -> Mostrar colunas específicas

**dados.describe()** -> Mostra um DataFrame com informações estatísticas 

**dados['Coluna'].mean()** -> Média de uma coluna específica

**dados.groupby('Coluna').mean()** -> Agrupar as informações de uma coluna específica e depois realizar a média ou outra operação

**dados.groupby('Coluna1')['Coluna2'].mean()** -> Agrupar as informações de uma coluna específica e depois realizar a média apenas de outra coluna específica

**dados.Coluna.unique()** -> Mostrar os dados de apenas uma coluna sem repetição

**dados.query('@variável in Coluna')** -> Seleciona informações lógicas dentro de um DataFrame

**dados.drop()** -> Remover linhas ou colunas com determinados valores

**dados.apply(função)** -> Aplicar uma função em uma coluna (Series)

**dados.applymap(função)** -> Aplicar uma função em mais de uma coluna (DataFrame)

**dados.duplicated()** -> Mostra os index que possuem valores duplicados

**dados.drop_duplicates()** -> Elimina as linhas que possuem valores duplicados

**df.to_csv('NomeDoArquivo', index=False)** -> Salva arquivo csv sem indexação

**df.to_excel('NomeDoArquivo', index=False)** -> Salva arquivo excel sem indexação

## Explorando os Dados

> Analisar os dados que foram importados

- **df.head(n)** -> Imprime as primeiras n linhas do dataframe

- **df['variável'].value_counts()** -> Conta o número de linhas com cada valor exclusivo da variável

- **df.shape** -> Tupla com o número de linhas e colunas do dataframe

- **df.columns** -> Mostra todas as colunas do dataframe

- **df.describe()** -> Estatísticas descritivas básicas para cada coluna do dataframe

- **df.quantile()** -> Retorna valores no quantil fornecido sobre o eixo solicitado

- **df.info()** -> Retorna informações sobre o dataframe

## Dados Duplicados

> Limpeza dos dados duplicados

- **df.duplicated().sum()** -> Verificar a quantidade de dados duplicados

- **df[df.duplicated()]** -> Filtra quais são as linhas duplicadas no dataframe

- **df.drop_duplicates()** -> Remove os dados duplicados

## Dados Nulos (NaN)

> Limpeza dos dados nulos (Not a Number)

- **df.isna().sum()** -> Quantidade de valores nulos em cada coluna

- **df.fillna(valor)** -> Substitue todos os dados NaN por algum valor

- **df.dropna()** -> Remove as amostras que contenham valores nulos em qualquer coluna do dataframe

- **df.interpolate()** -> Interpolar os valores NaN com os valores adjacentes (média)

## Variáveis Categóricas

> Binárias ou não binárias

- **df.replace()** -> Substitue valores por outros valores (0,1)

- **pd.get_dummies(df, dtype=int)** -> Aplica a técnica de One Hot Encoder (dummy) no dataframe

## Outliers

> Tratamento de Outliers

- **Q1 = df[coluna].quantile(.25)**

- **Q3 = df[coluna].quantile(.75)**

- **IIQ = Q3 - Q1**

- **limite_inferior = Q1 - 1.5 * IIQ**

- **limite_superior = Q3 + 1.5 * IIQ**

- **outliers_index = (df[coluna] < limite_inferior) | (df[coluna] > limite_superior)**

- **df[outliers_index]**

## Conversão de Valores

> Lembrar de converter tudo para valores numéricos

- **df['coluna'].astype(int)** -> Converter para inteiro, float, etc

- **df['coluna'].astype(np.int64)** -> Precisa importar a biblioteca numpy as np

- **df[df['coluna'] != '']** -> Fazer um novo dataframe com todas as linhas da coluna que não têm valores vazios

## Tratamento de Strings

> Tratar strings em dataframes

- **df['coluna'] = df.coluna.str.replace('(', '', regex=False).str.replace(')', '', regex=False)** -> Remover os parênteses

- **df['coluna'] = df['coluna'].str.upper()** -> Colocar tudo em maiúsculo ou minúsculo(lower)

Na biblioteca pandas, é possível fazer uma mudança do formato wide para o formato long com o uso do método melt(). A sintaxe para utilizar o método melt() é a seguinte:


O parâmetro id_vars é utilizado para identificar os nomes das colunas que serão mantidas no momento da transformação. Já o parâmetro value_vars é utilizado para identificar as colunas que serão transformadas nas colunas principais do formato long, uma contendo as categorias com os nomes das colunas e outra contendo os valores. O parâmetro var_name é usado para nomear a coluna contendo as categorias e o parâmetro value_name para nomear a coluna com os valores numéricos.



- Q1 -> Quarti 25%
- Q2 ou Mediana -> 50%
- Q3 -> 75%
- IIQ -> Q3-Q1 (Intervalo Interquartil)
- Limite Inferior -> Q1 - 1,5 x IIQ
- Limite Superior -> Q3 + 1,5 x IIQ
- Outliers -> Acima dos Limites

````python
import seaborn as sns
````

````python
# Boxplot para identificar outliers
sns.boxplot(x=df_sem_nulo['cliente.tempo_servico']);
````

````python
# Calculo para definir os limites
Q1 = df_sem_nulo['cliente.tempo_servico'].quantile(.25)
Q3 = df_sem_nulo['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5*IQR
limite_superior = Q3 + 1.5*IQR
````

````python
# Resgatar os indices dos outliers
outliers_index = (df_sem_nulo['cliente.tempo_servico'] < limite_inferior) | (df_sem_nulo['cliente.tempo_servico'] > limite_superior)
````
